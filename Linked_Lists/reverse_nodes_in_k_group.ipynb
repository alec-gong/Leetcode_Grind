#### Problem
Given the `head` of a linked list, reverse the nodes of the list `k` at a time, and return the modified list.

k is a positive integer and is less than or equal to the length of the linked list. If the number of nodes is not a multiple of k then left-out nodes, in the end, should remain as it is.

You may not alter the values in the list's nodes, only nodes themselves may be changed.

### Solution 1: Brute Force
We break down the linked list to linked lists of size k or smaller, put them in a list, and merge them together.

```python
def reverseKGroup(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
    def reverse(head):
        prev = None
        while head:
            temp = head.next
            head.next = prev
            prev = head
            head = temp
        return prev
    
    def mergeKLists(lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        def mergeTwoListNode(l1, l2) -> ListNode:
            dummy = ListNode()
            temp = dummy
            while l1:
                temp.next = l1
                l1 = l1.next
                temp = temp.next
            while l2:
                temp.next = l2
                l2 = l2.next
                temp = temp.next
            return dummy.next
        
        while len(lists) > 1:
            merged = []
            for i in range(0, len(lists), 2):
                list1 = lists[i]
                list2 = lists[i+1] if (i + 1) < len(lists) else None
                merged.append(mergeTwoListNode(list1, list2))
                
            lists = merged
        return lists[0] if lists else None
    
    count = 0
    lists = []
    dummy = ListNode()
    current = head
    
    # break down linked list
    while current:
        if count == 0:
            dummy = current
        if count + 1 == k:
            nextNode = current.next
            current.next = None
            lists.append(reverse(dummy))
            print(lists[-1])
            current = nextNode
            count = 0
        else:
            current = current.next
            count += 1
    if nextNode:
        lists.append(nextNode)
    # merge
    return mergeKLists(lists)
```

### Solution 2
To understand the solution, we have to first understand the meaning of the pointers. We divide the linked list by length of k, and for each group, we have before it a pointer `groupPrev`, and after it `groupNext`. 

So when we start reversing the list, the head is the node after `groupPrev`, and the end (which is also the first node of the reversed list) is `kth`, aka `groupNext.next`.

After reversing a group, `groupPrev` cannot no longer have a next reference to the head of the original group (which is now the tail of the reversed group). Instead, we should set the reference to the now head of the group, `kth`. Afterwards, we have to update `groupPrev` for next group, and set it to its original next reference.

```python
def reverseKGroup(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
    # need dummy because we potentially need to modify the first node
    dummy = ListNode(0, head)
    groupPrev = dummy
    
    while True:
        kth = self.getKth(groupPrev, k)
        # if less than k nodes, no need to reverse
        if not kth:
            break
        groupNext = kth.next

        # reverse gropu
        prev, curr = kth.next, groupPrev.next
        while curr != groupNext:
            temp = curr.next
            curr.next = prev
            prev = curr
            curr = temp
        
        # update groupPrev
        temp = groupPrev.next
        groupPrev.next = kth
        groupPrev = temp
    
    return dummy.next

def getKth(self, head, k):
    while head and k > 0:
        head = head.next
        k -= 1
    return head
```
